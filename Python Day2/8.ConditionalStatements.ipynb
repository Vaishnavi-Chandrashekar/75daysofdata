

-> IF Statement in python

In Python, the `if` statement is used for conditional execution. It allows you to execute a block of code only if a specified condition is true. Here's the basic syntax:
    
    
    if condition:
        # code block to execute if condition is true
    


```python
 #if statement
 
age = 24
if age>=20:
    print('age is greater than 20')
```

```python
if True:
    print('true')
```

 IF-ELSE Statement in python
In Python, the `if`-`else` statement allows you to execute one block of code if a condition is true and another block of code if the condition is false. Here's the basic syntax:
    
    
    if condition:
        # code block to execute if condition is true
    else:
        # code block to execute if condition is false
    
    

You can also extend the `if` statement with `elif` (short for "else if") and `else` to handle multiple conditions:
    
    
    if condition1:
        # code block to execute if condition1 is true
    elif condition2:
        # code block to execute if condition2 is true
    else:
        # code block to execute if none of the above conditions are true


```python
	x = 10
	if x > 5:
	    print("x is greater than 5")
	else:
    print("x is less than or equal to 5")
```

```python
#else: the else condition works if the "if" condition is false
if True:
    print('true')
else:
    print('false')
```

```python
#else: the else condition works if the "if" condition is false
if False:
    print('true')
else:
    print('false')
```

```python
age = int(input('enter your age: '))

if age>=18:
    print('you are eligible to vote')
else:
    print('you are under 18')
```

```python
	x = 10
	if x > 5:
	    print("x is greater than 5")
	elif x == 5:
	    print("x is equal to 5")
	else:
	    print("x is less than 5")
```


->    `Elif ladder`
In Python, there isn't a specific "else if" construct like in some other languages (e.g., C, C++, Java), but you can achieve the same functionality using `elif`, which stands for "else if".
    
Here's how you can use it in a "ladder" style to handle multiple conditions:
    
    
    if condition1:
        # code block to execute if condition1 is true
    elif condition2:
        # code block to execute if condition2 is true
    elif condition3:
        # code block to execute if condition3 is true
    # Add more elif blocks as needed
    else:
        # code block to execute if none of the above conditions are true
    
    
Each `elif` block is evaluated only if the preceding `if` or `elif` conditions are false. If any of the conditions are true, the corresponding block of code is executed, and the rest of the `elif` and `else` blocks are skipped.
    


```python
	x = 10
	if x > 100:
	    print("x is greater than 100")
	elif x > 50:
	    print("x is greater than 50 but not greater than 100")
	elif x > 10:
	    print("x is greater than 10 but not greater than 50")
	else:
	    print("x is 10 or less")
```

```python
#elif: allows to check for multiple conditiions.

age = int(input('enter your age'))

if age<=13:
    print('you are a child')
elif 13<age<18:
    print('you are a teen')
else:
    print('you are are an adult')
    
```

->  `Nested if`
    Nested `if` statements in Python allow you to have one `if` statement inside another. This can be useful when you need to check for multiple conditions, and the actions to take depend on the outcomes of those conditions. Here's the basic syntax:
    
    
    if condition1:
        # code block to execute if condition1 is true
        if condition2:
            # code block to execute if both condition1 and condition2 are true
        else:
            # code block to execute if condition1 is true but condition2 is false
    else:
        # code block to execute if condition1 is false


```python
	x = 10
	if x > 5:
	    print("x is greater than 5")
	    if x < 15:
	        print("x is also less than 15")
	    else:
	        print("x is not less than 15")
	else:
	    print("x is less than or equal to 5")
```

In this example, there's a nested `if` statement inside the outer `if` statement. It first checks if `x` is greater than 5. If it is, it proceeds to check if `x` is also less than 15. Depending on the outcome of these conditions, different blocks of code are executed.
    
You can have multiple levels of nesting, but it's essential to maintain readability and avoid overly complex structures if possible.


```python
#nested conditional statements
num = int(input('enter a number'))

if num > 0:
    print('the number entered is positive')
    if num%2 == 0:
        print('its an even number')
    else:
        print('its an odd number')
else:
    print('the number entered is either 0 or negative')
```

```python
#nested conditional statements
num = int(input('enter a number'))

if num > 0:
    print('the number entered is positive')
    if num%2 == 0:
        print('its an even number')
    else:
        print('its an odd number')
elif num<0:
    print('the number is negative')
else:
    print('the number entered is 0')
```

```python
#determine if the given year is a leap year

year = int(input('enter year'))

if year%4 == 0:
    if year%100 == 0:
        if year%400 == 0:
            print(year, 'is a leap year')
        else:
            print(year, 'is not a leap year')
    else: 
        print(year,'is a leap year')
else:
    print(year, 'is not a leap year')
```
